We have a really low data before 1980, so we are going to erase this data and then proceed to the pre-processing stage.

```python
df=df[df['Year'] > 1980]
```

```python
countrys = df['Entity'].unique()

df_pre_cleaned = pd.DataFrame()

for country in countrys:
    df_country = df[df['Entity'] == country].copy()

    # Remove "Entity" column to do the interpolation
    df_country_temp = df_country.drop('Entity', axis=1)

    # Interpolate the data and fill the NaN values with the mean data of each country
    df_country_cleaned_temp = df_country_temp.interpolate().fillna(df_country_temp.mean())

    # Add the "Entity" column again
    df_country_cleaned_temp.insert(0, 'Entity', country)

    # Create a new dataframe from the original
    df_pre_cleaned = pd.concat([df_pre_cleaned, df_country_cleaned_temp])

# Ordenar o DataFrame resultante
df_pre_cleaned.reset_index(drop=True)
df_pre_cleaned.describe()
```

```python
df_pre_cleaned.isnull().sum()
```

We can see a big amount of data been missed and the reason is really simple to notice, there is population data missing, this have direct influency on "per capita" values. All that is due to a whole missing data about population in the 741 rows represents around 8% of the data. Let's see which countrys will be erased if we drop all these data.

```python
df_cleaned = df_pre_cleaned.dropna()
df_cleaned.describe()
```

```python
country2 = df_cleaned['Entity'].unique()
paises_nao_presentes = set(countrys) - set(country2)

print("Countries that have been deleted from the original dataframe: \n", pd.DataFrame(paises_nao_presentes))
```

25 countrys will be excluded from this analysis if we decide to drop the data. Most of these countrys are islands (or even countrys that doesn't exist anymore). To deal with this problem I decided to split into 2 datasets, the first one is excluding all the data with population (including "per capita" values), the other one will be just with the data cleaned and ignoring these 25 countrys.

###3.1. Preparing the first dataframe + world dataframe

---
Expanding upon our analysis, we can leverage the existing data to engineer new features that offer deeper insights into the economic dynamics of countries. For instance, introducing a metric like "Gas Surplus" can serve as a valuable proxy for estimating a country's import and export activities related to gas commodities. This derived data not only enriches our understanding of economic patterns but also enhances the scope of our EDA.

By quantifying "Gas Surplus," we effectively capture the disparity between a nation's gas production and consumption levels. This information provides nuanced insights into the country's energy self-sufficiency and its reliance on global trade networks for meeting energy demands.


```python
#First we remove the population data
df_pre_cleaned_nopop = df_pre_cleaned.iloc[:, :8]
#Then we merge dataframes
df_nopop = pd.merge(df2,df_pre_cleaned_nopop, on='Entity', how='inner')
country3 = df_nopop['Entity'].unique()#To check which countrys the dataframe didn't catch
df_nopop=df_nopop.dropna()
df_nopop["Gas Surplus(m³)"] = df_nopop["Gas production(m³)"]-df_nopop["Gas consumption(m³)"]
df_nopop["Coal Surplus(Ton)"] = df_nopop["Coal production(Ton)"]-df_nopop["Coal consumption(Ton)"]
df_nopop["Oil Surplus(m³)"] = df_nopop["Oil production(m³)"]-df_nopop["Oil consumption(m³)"]
df_nopop.describe()
```

```python
paises_nao_presentes = set(countrys) - set(country3)
print("Countrys missing from pre cleaned to no population dataset:", paises_nao_presentes)
```

As you can see, the `df2` was pre treated and we only don't get the "World and U.S. Territories" data. Since a lot of specific US territories have been spread in this dataset, the U.S territories will be ignored. The data "world" will be on a different dataframe for analysis purposes.

```python
world_df = df_cleaned[df_cleaned['Entity'] == "World"].copy().reset_index(drop=True)
world_df.describe()
```

Now we have 2 dataframes ready, but we want to prepare the dataset with the population so we can use it on the Power BI analysis.

```python
df_pop = df_pre_cleaned.iloc[:, [0, 1, 8,9,10,11,12,13,14]]
df_pop = pd.merge(df2,df_pop, on='Entity', how='inner')
df_pop=df_pop.dropna().reset_index(drop=True)
df_pop.describe()
```

Now we have 3 dataframes ready to be analyzed. A briefly description of each dataframe:

*   `df_nopop` This dataframe have all the comsuption/production rate but doesn't have any rate correlated to the population. This one, have more data than the `df_pop`, mosltly islands.

###Extra  



---



Since we are going to do this to all the types of energy, let's create functions so we can save some lines of code.

```python
def rank_energy(df,energy,n_rank,ascending):

  df_top_gas_prod = df.groupby('Entity').apply(lambda x: x.nlargest(1, energy)).reset_index(drop=True)

  df_top_gas_prod=df_top_gas_prod[df_top_gas_prod[energy] > 0]

  df_top_gas_prod = df_top_gas_prod.sort_values(by=energy, ascending=ascending).head(n_rank)

  top_country = df_top_gas_prod["Entity"]

  df_top_gas_prod = df[df['Entity'].isin(top_country)]

  return df_top_gas_prod

def plot_cons_prod(df,energy,axis):

  sns.lineplot(x='Year', y=energy, hue='Entity', data=df, marker='o', markersize=8,ax=axis)
  axis.set_title(str(energy)+' by year')
  axis.set_xlabel('Year')
  axis.set_ylabel(energy)
  axis.legend(loc='upper left')
```
